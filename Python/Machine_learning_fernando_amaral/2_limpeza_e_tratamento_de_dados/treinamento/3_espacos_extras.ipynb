```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Cidade': ['São Paulo  ', 'Rio de Janeiro ', 'Belo Horizonte ']})

df
```

1. Remover os espaços extras

```python
df['Cidade'] = df['Cidade'].str.strip()
df
```

2. Converter tudo para MAIÚSCULO

```python
df['Cidade'] = df['Cidade'].str.upper()
df
```

3. Converter tudo para minúsculo

```python
df['Cidade'] = df['Cidade'].str.lower()
```
