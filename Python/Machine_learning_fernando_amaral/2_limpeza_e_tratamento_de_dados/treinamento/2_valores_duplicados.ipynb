### Problema 2 — Dados Duplicados

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'cliente':['A', 'B', 'B' ,'C', 'D', 'D'],
    'valor':[10,20,20,30,40,40]
})
```

1. Identificar registros duplicados

Crie uma coluna temporária ou mostre quais linhas são duplicadas.

```python
df['linhas_duplicadas'] = df.duplicated()
df
```

2. Contar quantos duplicados existem

Retorne apenas a quantidade.

```python
df.duplicated(subset=['cliente']).sum()
```

3. Remover os registros duplicados

Crie um novo DataFrame sem duplicados.

```python
df = df.drop(columns=['linhas_duplicadas'])
```

```python
df
```

```python
df = df.drop_duplicates()
```

4. Mostrar o DataFrame final

```python
df
```
