### Problema 1 — Valores Nulos

Crie o DataFrame:

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'nome':['Ana', 'Pedro', np.nan, 'Maria'],
    'idade':[20, np.nan, 35, 40]
})
```

1. Mostrar quantidade de nulos por coluna.

```python
df.isnull().sum()
```

2. Mostrar porcentagem de nulos.

```python
percrentual_nulos = (df.isnull().sum().sum()/df.size)*100
percrentual_nulos
```

3. Substituir idade nula pela média.

```python
df['idade'] = df['idade'].fillna(df['idade'].mean())
df
```
