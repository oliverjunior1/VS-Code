### 1. Análise de clientes

```python
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carla", "Diego", "Eva"],
    "idade": [22, 35, 29, 41, 30],
    "cidade": ["São Paulo", "Rio de Janeiro", "São Paulo", "Recife", "Curitiba"],
    "compras": [3, 8, 5, 2, 10],
}

df = pd.DataFrame(dados)
```

2. Filtre clientes de ``São Paulo``.

```python
clientes_de_SP = df[df['cidade'] == 'São Paulo']
clientes_de_SP
```

3. Filtre clientes com mais de ``5`` compras.

```python
clientes_5_ou_mais_compras = df[df['compras']>=5]
clientes_5_ou_mais_compras
```

4. Calcule a média de idade.

```python
media_idade = df['idade'].mean()
media_idade
```

### 2. Ordenação e novas colunas

1. Ordene os clientes por ``compras``, do maior para o menor.

```python
df_ordenado_compras = df.sort_values(by='compras', ascending=False)
df_ordenado_compras
```

2. Crie uma coluna chamada ``cliente_frequente``.

```python
import numpy as np
df['cliente_frequente'] = np.where(
    df['compras']>5,
    'Frequente',
    'Não frequente'
)

df
```
