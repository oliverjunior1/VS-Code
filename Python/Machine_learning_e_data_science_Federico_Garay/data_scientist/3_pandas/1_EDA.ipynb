### Criando uma Base

```python
import pandas as pd

dados = {
    'Nome': ['Ana', 'Pedro', 'Júlio', 'Carlos', 'Maria'],
    'Idade': [25, 30, 22, 40, 35],
    'Salário':[3000,5000,4000,7000,6000]
}

df = pd.DataFrame(dados)
df
```

# Parte 5 — Filtrando Dados

Salário acima de 5000:

```python
df[df['Salário'] > 5000]
```

Idade menor que 30:

```python
df[df['Idade'] < 30]
```

Duas condições maior que 25 e maior que 4000:

```python
df[
    (df['Salário'] > 4000) & (df['Idade'] > 25)
]
```

Duas condições maior que 30 ou primeira letra do nome M:

```python
df[
    (df['Nome'] > 'A*') | (df['Idade'] > 30)
]
```
