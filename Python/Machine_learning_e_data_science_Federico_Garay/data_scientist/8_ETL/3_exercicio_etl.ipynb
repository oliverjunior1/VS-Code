### Questão 5 - Transform

Crie uma nova coluna chamada:

```python
import numpy as np

condicoes = [
    vendas['valor'] >= 3000,
    vendas['valor'] >= 1000
]

categorias = [
    'Premium',
    'Intermediaria'
]

vendas['categoria'] = np.select(
    condicoes,
    categorias,
    default='Basica'
)
```
