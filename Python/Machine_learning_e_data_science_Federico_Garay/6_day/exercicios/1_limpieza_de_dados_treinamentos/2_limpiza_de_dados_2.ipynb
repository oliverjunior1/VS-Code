Primeiro, crie este DataFrame:

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'nome': ['Ana', 'Bruno', np.nan, 'Carlos', 'Ana'],
    'idade': [25, np.nan, 30, 150, 25],
    'salario': [3000, 4000, np.nan, 5000, 3000]
})

df
```

### Questão 1

Como descobrir quantos valores nulos existem em cada coluna?

```python
df.isnull().sum()
```

### Questão 2

Como visualizar apenas as linhas que possuem valores nulos?

```python
df[df.isnull().any(axis=1)]
```

### Questão 3

Como substituir os valores nulos da coluna idade pela média das idades?

```python
age_medium = df.idade.mean()
age_medium
```

```python
df.idade = df.idade.fillna(age_medium)
df.idade
```
