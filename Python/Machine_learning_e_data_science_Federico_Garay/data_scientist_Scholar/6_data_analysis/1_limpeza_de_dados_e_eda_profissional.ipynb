### Limpeza de Dados e EDA Profissional

Uma regra muito conhecida em Data Analytics é:

- 80% do tempo = preparando dados
- 20% do tempo = analisando dados

```python
import pandas as pd
```

```python

dados = {
    "nome": ["Ana", "Pedro", "Julia", "Pedro", None],
    "idade": [25, 30, None, 30, 40],
    "salario": [3000, 5000, 4000, 5000, None]
}

df = pd.DataFrame(dados)
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Passo 2 — Encontrar Valores Nulos

```python
df.isnull().sum()
```

Tratando Valores Nulos

Opção 1 — Remover

```python
df = df.dropna()
df
```

### Opção 2 — Preencher

Idade média:

```python
df['idade'] = df['idade'].fillna(
    df['idade'].mean()
)

df
```

```python
df['salario'] = df['salario'].fillna(
    df['salario'].mean()
)
```

### Passo 3 — Encontrar Duplicatas

Verificar:

```python
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

### Passo 4 — Tipos de Dados

Verificar:

```python
df.dtypes
```

```python
df['idade'] = df['idade'].astype(int)
```

### Passo 5 — Estatísticas Importantes

Média:

```python
df['salario'].mean()
```

mediana

```python
df['salario'].median()
```

Desvio padrão:

```python
df['salario'].std()
```

Máximo:

```python
df['salario'].max()
```

Mínimo

```python
df['salario'].min()
```

### Passo 6 — Filtragem Profissional

Salários acima da média:

```python
df[df['salario']>df['salario'].mean()]
```

Funcionários acima de 30 anos:

```python
df[df['idade']>30]
```

```python
df[
    (df['idade']>25) &
    (df['salario']>4000)
]
```

### Passo 7 — Criar Novas Colunas

Salário anual:

```python
df['salario_anual'] = df['salario'] * 12
```

Faixa salarial:

```python
df["categoria"] = df["salario"].apply(
    lambda x: "Alta" if x > 5000 else "Normal"
)
```

### Agrupamentos

Tabela:

```python
dados = {
    "departamento": ["TI","TI","RH","RH"],
    "salario": [5000,7000,4000,6000]
}

df = pd.DataFrame(dados)
```

Média por departamento:

```python
df.groupby('departamento')['salario'].mean()
```

Total por departamento:

```python
df.groupby('departamento')['salario'].sum()
```
