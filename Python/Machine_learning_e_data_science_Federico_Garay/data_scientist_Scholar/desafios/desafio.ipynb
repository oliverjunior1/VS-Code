1. Receba 5 notas.

```python
x = 50,100,80,45,75
```

2. Armazene em uma lista.

```python
lista = list(x)
```

3. Calcule:
- média
- maior nota
- menor nota

```python
media = sum(lista) / len(lista)
maior_nota = max(lista)
menor_nota = min(lista)
print(f'média: {media}, maior nota: {maior_nota}, menor nota: {menor_nota}')
```

4. Informe se a turma foi:
- Excelente (média ≥ 9)
- Boa (média ≥ 7)
- Regular (média ≥ 5)
- Fraca (média < 5)

```python
match media:
    case media if media >= 90:
        print('Excelente')
    case media if media >= 70:
        print('Boa')
    case media if media >= 50:
        print('Regular')
    case _:
        print('fraca')
```
