### Exemplo simples de Árvores de decisão

```python
from sklearn import tree
```

- Dados de exemplo (features)
- Cada linha representa [peso, textura]
- peso:1 = leve, 2 = pesado
- textura: 1 = áspera, 2 = lisa

```python
X = [[1,1],[2,1],[2,2],[1,2]]
```

# Classes (labels)

0 = maçã, 1 = laranja

```python
y = [0,1,1,0]
```

Criando o classificador

```python
clf = tree.DecisionTreeClassifier()
```

Treinando o modelo

```python
clf = clf.fit(X,y)
```

- Fazendo uma previsão
- Exemplo: fruta com peso=2 e textura=2

```python
predicao = clf.predict([[2,2]])
```

```python
print('Predicao: ', predicao) 
```

- `x` são atributos (peso e textura).

- `y` são as classes (maçã ou laranja).

- O modelo aprende a separar os dados com base nesses atributos.

- Depois, usamos `predict` para prever a classe de uma nova entrada.
