### Limpeza de dados

```python
import pandas as pd

data = {'Id_producto':[1001,1002,1003,1003],
        'Cantidad_vendida':[30, None, 25,25],
        'Precio':[20.5,15.0,None, 22.5]}

df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df_eliminados = df.dropna()
df_eliminados
```

```python
valores_nuevos = {'Candidad_vendida':0,"Precio":df['Precio'].mean()}
df_rellenados = df.fillna(valores_nuevos)
df_rellenados
```

```python
df_rellenados['Cantidad_vendida'] = df_rellenados['Cantidad_vendida'].astype('Int64')
df_rellenados
```

```python
df_unicos = df_rellenados.drop_duplicates()
df_unicos
```

```python
df_unicos = df_rellenados.drop_duplicates(subset='Id_producto')
df_unicos
```
