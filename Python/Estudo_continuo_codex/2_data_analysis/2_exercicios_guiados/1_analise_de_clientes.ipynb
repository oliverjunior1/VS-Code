- importando pandas

```python
import pandas as pd
```

- dataset e dataframe

```python
dados = {
    "nome": ["Ana", "Bruno", "Carla", "Diego", "Eva"],
    "idade": [22, 35, 29, 41, 30],
    "cidade": ["São Paulo", "Rio de Janeiro", "São Paulo", "Recife", "Curitiba"],
    "compras": [3, 8, 5, 2, 10],
}

df = pd.DataFrame(dados)
```

1. Mostre apenas as colunas ``nome`` e ``cidade``.

```python
df.groupby('nome')['cidade'].unique()
```

2. Filtre clientes de ``São Paulo``.

```python
Sao_paulo = df[df['cidade'] == 'São Paulo']
Sao_paulo
```

3. Filtre clientes com mais de ``5`` compras.

```python
mais_de_cinco = df[df['compras']>5]
mais_de_cinco
```

4. Calcule a média de idade.

```python
df['idade'].mean()
```
