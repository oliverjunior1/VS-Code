### Filtrando e resumindo

- importando pandas

```python
import pandas as pd
```

- Criando o dataset

```python
dados = {
    "produto": ["Mouse", "Teclado", "Caderno", "Caneta", "Monitor"],
    "categoria": ["Informática", "Informática", "Papelaria", "Papelaria", "Informática"],
    "preco": [50, 120, 25, 5, 900],
    "quantidade": [2, 1, 4, 10, 1],
}
```

- transformando em dataset utilizavel por pandas

```python
df = pd.DataFrame(dados)
```

- criando uma nova coluna

```python
df['total_venda'] = df['preco'] * df['quantidade']
```

```python
df
```

- filtrando parecido com where em sql

```python
informatica = df[df['categoria']=='Informatica']
informatica
```

- Estatisticas simples

```python
'Preço médio', df['preco'].mean()
```

```python
'Produto mais caro:', df['preco'].max()
```

```python
'Total vendido:', df['total_venda'].sum()
```

- Agrupamento, parecido com GROUP BY em SQL

```python
resumo = df.groupby('categoria')['total_venda'].sum()
resumo
```
