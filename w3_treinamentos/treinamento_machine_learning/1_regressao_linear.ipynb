### 🍰 Passo 1 — Preparar o ambiente

```python
import numpy as np
```

- Para que serve:  
``NumPy`` é a biblioteca que usamos para trabalhar com vetores e matrizes de forma eficiente.
Machine Learning vive em cima disso.

### 🍰 Passo 2 — Criar o dataset

```python
# Tamanhos das casas (em m³)
X = np.array([50,80,120,200], dtype=float)

# Preços das casas (em milhares de reais)
y = np.array([150,210,300,450], dtype=float)
```

- X: é nosso vetor de entrada (features) → tamanho da casa.

- y: é nosso vetor de saída (target) → preço da casa.

- dtype=float: garante que estamos trabalhando com números reais (importante para cálculos).

### 🍰 Passo 3 — Inicializar os parâmetros do modelo

Nosso modelo é: ``y = w.x + b``

```python
# Parâmetros do modelo (pesos)
w = 0.0 #Peso inical
b = 0.0 # viés (bias) inicial

# Taxa de aprendizado
learning_rate = 0.0001

# Número de épocas (quantas vezes vamos treinar)
epochs = 10000
```

- w: controla o quanto o tamanho da casa influencia o preço.

- b: é o valor base, mesmo se o tamanho fosse zero.

- learning_rate: quão grande é o passo de atualização dos parâmetros.

- epochs: quantas vezes vamos repetir o processo de aprendizado.

### 🍰 Passo 4 — Definir a função de previsão

```python
def predict(x):
    return w * x + b
```

- Para que serve:  
Essa função representa o modelo de regressão linear.
Dado um x (tamanho da casa), ela retorna ŷ (preço previsto).

### 🍰 Passo 5 — Definir a função de erro (MSE)

```python
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)
```

- Para que serve:  
Mede o quão ruim estão as previsões.
Quanto menor o MSE, melhor o modelo.

### 🍰 Passo 6 — Treinamento (Gradient Descent)

```python
for epoch in range(epochs):
    y_pred = predict(X)

    error = y_pred - y

    dw = (2/len(X)) * np.dot(error, X)
    db = (2/len(X)) * np.sum(error)

    w -= learning_rate * dw
    b -= learning_rate * db

    if epoch % 1000 ==0:
        mse = mean_squared_error(y, y_pred)
        print(f"Epoch {epoch}, MSE: {mse:.2f}, w: {w:.4f}, b: {b:.4f}")
```
